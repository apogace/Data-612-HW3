1. Create three different meaningful charts using Seaborn package on your selected data set.

```
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt
df = sns.load_dataset("penguins")
df
```

```
df.value_counts('species'),df.value_counts('island')
```

```
count, ax = plt.subplots()

ax = sns.countplot('sex', data=df)

ax.set_title('Penguine Sex')

ax.set_ylabel('Frequency')

plt.show()
```

```
scatter, ax = plt.subplots()

ax = sns.regplot(x='body_mass_g', y ='flipper_length_mm', data=df)

ax.set_title('Relationship between flipper length and body mass')

ax.set_xlabel('Body Mass(g)')

ax.set_ylabel('Flipper Length(mm)')

plt.show()
```

```
scatter, ax = plt.subplots()

ax = sns.regplot(x='body_mass_g', y ='bill_length_mm', data=df)

ax.set_title('Relationship between bill length and body mass')

ax.set_xlabel('body mass (g)')

ax.set_ylabel('bill length (mm)')

plt.show()
```

```
SP1 = df[df['species'] == 'Adelie']
SP2 = df[df['species'] == 'Gentoo']
SP3 = df[df['species'] == 'Chinstrap']
```

```
fig = plt.figure()

axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)

axes1.hist(df['body_mass_g'], bins=20)
axes2.hist(SP1['body_mass_g'], bins=20)
axes3.hist(SP2['body_mass_g'], bins=20)
axes4.hist(SP3['body_mass_g'], bins=20)

axes1.set_title("Body Mass Histograph of all Penguines")
axes2.set_title("Body Mass Histograph of Adelie")
axes3.set_title("Body Mass Histograph of Gentoo")
axes4.set_title("Body Mass Histograph of Chinstrap")

fig.set_size_inches(12, 10)

fig.show()
```
